- the minion game

```python
def minion_game(string):
    vowels = 'AEIOU'
    stuart = int()
    kevin = int()
    for i in range(len(string)):
        if string[i] in vowels:
            kevin += (len(string) - i)
        else:
            stuart += (len(string) - i)
    if stuart > kevin:
        print(f'stuart {stuart}')
    elif kevin > stuart:
        print(f'kevin {kevin}')
    else:
        print('Draw')
```

```python
def minion_game(string):
    vowel = [len(string) - x.start() for x in list(re.finditer('[AEIOU]', string))]
    consonant = [len(string) - y.start() for y in list(re.finditer('[^AEIOU]', string))]
    kevin = sum(vowel)
    stuart = sum(consonant)
    
    if kevin > stuart:
        print(f'kevin {kevin}')
    elif stuart > kevin:
        print(f'stuart {stuart}')
    else:
        print('Draw')
```
